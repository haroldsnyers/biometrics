# Jupyter Notebook example

```python
import scipy
import numpy as np

from scipy import misc, signal
from matplotlib import pyplot as plt
```

## Loading an image
First we load the image we wish to apply the filters to. For this session we'll make use of an example image from SciPy. 

```python
img = misc.face()
plt.imshow(img)
plt.show()
```

## Create a filter
<p> As the name spells out gaussian noise is noise that is gaussian-distributed </p>

$$
    X \sim \mathcal{N(\mu,\sigma^2)}.
$$

<p>We'll clip the noise to be able to visualise it. However, notice that the filter contains both positive and negative values.</p> 

```python
sigma_sq, mu = 75, 50
# * -> convert tuple to integer
noise = (sigma_sq*np.random.randn(*img.shape) + mu).astype(np.int32)
plt.imshow(np.clip(noise,0,255))
plt.show()
```

## Adding gaussian noise
Now that we have a gaussian filter we simply add the filter to the image and clip it to make sure the values remain within the pixel domain.

```python
noisy_image = np.clip(img + noise,0,255)
plt.imshow(noisy_image);
```

## Gaussian Smoothing
<p>To reduce noise in an image, one often makes use of gaussian blurring. This means that we create a filter, or so-called kernel, that we slide or convolve across the image. </p>
<p>Try to create a Gaussian smoothing kernel and apply is to the noisy_image. Tip: you can make use of the scipy.ndimage.gaussian_filter module or implement it yourself using scipy.signal.convolve2d. 
Remember, your kernel is an image too. You can visualize it in the same way as you did before.</p>

```python
from scipy.ndimage import gaussian_filter
fig = plt.figure(figsize=(15,15))
plt.gray()  # show the filtered result in grayscale
ax1 = fig.add_subplot(131)  # left side
ax2 = fig.add_subplot(132)  # middle side
ax3 = fig.add_subplot(133)  # right side

result = gaussian_filter(noisy_image, sigma=5)

ax1.imshow(img)
ax2.imshow(noisy_image)
ax3.imshow(result)
plt.show()
```
